# EMPLOYEE ATTRITION PREDICTION USING LOGISTIC REGRESSION MODEL


# The Aim of the project is to apply data science techniques using supervised machine learning approach with Logistic regression model to predict the likelihood that an employee will either leave or stay in a company. 

```python
##Import all neccesary Libraries for the Machine Learning Programming (Numpy,Pandas,Sklearn - Sklearn is essential for 
#Machine learning Model)

import numpy as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
```

```python
 #The data will be split into 70% to train the machine to learn patterns in order to predict the outcome.
#the Data will be split into 30% to test the machine after it has learned the patterns. we shall use test to know how accurate the machine can predict the outcome
#Our goal is to succesfuly predict the an outcome given the data. The outcome in the dataset is a columnn in the dataset called  ('Status')


train = pd.read_csv('train.csv') #Initialize a Variable for to read Train Data set
test = pd.read_csv('test2.csv')  #Initialize a Variable for to read Test Data set
```

```python
train.rename(columns={'Emp ID':'EmpID'},inplace=True) #Rename the Employee I.D coloumn of Train Data to 'EmpID'
```

```python
test.rename(columns={'Emp ID':'EmpID'},inplace=True)  #Rename the Employee I.D coloumn of test Data to 'EmpID'
```

```python
#At this stage we shall write a code to create additional 10 columns for each unique department what we call one not encoding. 
##The idea is to return all binary positive values (1) for every time a row contains each department in both the train and test dataset
##The Initial Department column contains the Department listed per row. However, we will create unique 10 additional columns contains 
##all binary positive values (1) and negative value (0) for every time a row containing any department is referenced.


train = pd.get_dummies(train,columns=['dept'],drop_first=True) 
#This code will create a column per department and return positive binary or negative binary number in train dataset

test = pd.get_dummies(test,columns=['dept'],drop_first=True)   
#This code will create a column per department and return positive binary or negative binary number in train datase


#This code will also remove the last unique value column after splitting the unique values into serpate columns as part of the data set
```

```python
##At this stage we want to write a function to return rank numeric values for the three unique items in Salary column.\
##This function will return value 1 or any item in the salary column called 'Low', we will return 2 for 'medium', 
##This code will return 3 for 'High'

def rank(x):
    if x == 'low':
        return 1
    elif x == 'medium':
        return 2
    else:
        return 3
```

```python
train
```

```python
train['sal_status'] = train['salary'].apply(lambda x: rank(x)) 
#This code will create a new column called(New_Salary) in our train dataset and appy lambda to write a function over the initial Salary column.
#This code is written in order to return the rank (1,2 or 3) 

test['sal_status'] = test['salary'].apply(lambda x: rank(x))   
#This code will create a new column called(New_Salary) in our test dataset and appy lambda to write a function over the initial Salary column.
#This code is written in order to return the rank (1,2 or 3) 
train['new_satisfaction_level'] = train['satisfaction_level']*100
test['new_satisfaction_level'] = test['satisfaction_level']*100
train.drop(['satisfaction_level'],axis=1,inplace=True)
test.drop(['satisfaction_level'],axis=1,inplace=True)

train['new_last_evaluation'] = train['last_evaluation']*100
test['new_last_evaluation'] = test['last_evaluation']*100
train.drop(['last_evaluation'],axis=1,inplace=True)
test.drop(['last_evaluation'],axis=1,inplace=True)
```

```python
train.head() #we call our train variable which we read csv file into. it shows all changes applied in the previous lines of code
```

```python
train.drop(['salary'],axis=1,inplace=True)
#train.drop(['EmpID'],axis=1,inplace=True)
#This code will delete the Initial salary column train dataset because our machine can only read numbers not text

test.drop(['salary'],axis=1,inplace=True)
#test.drop(['EmpID'],axis=1,inplace=True)
#This code will delete the Initial salary column test dataset because our machine can only read numbers not text
```

```python
##At this stage we want to write a function to return rank numeric values for each unique years in YearOfRecruitment column.
##This function will return value 1 for Year 2017, value 2 for 2016 etc. until the end of the final year which is 2009.
##The function will return value 8


def rank_2(x):
    if x == 2017:
        return 0
    elif x == 2016:
        return 1
    elif x == 2015:
        return 2
    elif x == 2014:
        return 3
    elif x == 2013:
        return 4
    elif x == 2012:
        return 5
    elif x == 2011:
        return 6
    else: 
        return 8
```

```python
train['YOE'] = train['YearOfRecruitment'].apply(lambda x: rank_2(x))
#This Code will create a new column in our train dataset called(YOE) - Years Of Experience in our train dataset and appy lambda to write this function above over the initial Salary column.
#The function is applied in order to return the rank (8,7,6,5,4,3,2,1) 
```

```python
test['YOE'] = test['YearOfRecruitment'].apply(lambda x: rank_2(x))
#This Code will create a new column in our test dataset called(YOE) - Years Of Experience in our train dataset and appy lambda to write this function above over the initial Salary column.
#The function is applied in order to return the rank (8,7,6,5,4,3,2,1) 
```

```python
train.drop(['YearOfRecruitment'],axis=1,inplace=True)
test.drop(['YearOfRecruitment'],axis=1,inplace=True)
train.drop(['dept_accounting'],axis=1,inplace=True)
train.drop((['dept_management','dept_marketing','dept_product_mng','dept_sales','dept_support','dept_technical']),axis=1,inplace=True)
test.drop(['dept_accounting'],axis=1,inplace=True)
test.drop((['dept_management','dept_marketing','dept_product_mng','dept_sales','dept_support','dept_technical']),axis=1,inplace=True)
```

```python
train['intercept'] = 1
test['intercept'] = 1
```

```python
#At this state we shall create Two variable, X_Train and Y_Train.
#The purpose is to train our machine to learn the train dataset and also learn the outcomes ('Status') thereby making the machine to understand the patterns of the data itself and the target outcome ('Status') which is the desired result 
#Our goal is to succesfuly predict the an outcome given the data. The outcome in the dataset is a columnn in the dataset called  ('Status')


X = train.drop(['Status'],axis=1) 
#This code will delete only the Initial status column in the train dataset because staus is what we are trying to predict as our target
```

```python
y = train['Status']
#This code will return only the Initial status column because this we need the train the machine to learn the X train datses
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
algorithm = LogisticRegression() #LogisticRegression is the model we will adopt to train the data set
algorithm.fit(X_train,y_train) #This is a code to actually train the data set.
prediction = algorithm.predict(X_test) 
```

```python
accuracy_score(prediction,y_test) 
```
